```python
import seaborn as sns
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips_df=pd.read_csv("tips.csv")
```

```python
tips_df
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df)
plt.show()
```

```python
order=['Thur','Fri','Sat','Sun']
h_order=['Male','Female']
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,order=order,hue_order=h_order)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,estimator=np.max,errorbar=('ci',30),
            n_boot=2,orient='v')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,color='r')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,palette='hot'
           ,saturation=.5,err_kws={'color':'.5','linewidth':12},capsize=1)
plt.show()
```
