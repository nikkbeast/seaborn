```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tit_df=pd.read_csv('titanic.csv')
tit_df
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',data=tit_df,style='who',size='who')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='age',y='fare',hue='sex',data=tit_df,style='who',size='who',
               sizes=(100,400))
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='who',y='fare',hue='alive',data=tit_df,style='alive',size='who',
               sizes=(100,400),palette='hot')
plt.show()
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='age',y='fare',data=tit_df)
sns.lineplot(x='age',y='fare',data=tit_df)
sns.barplot(x='age',y='fare',data=tit_df)
```
