```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gw=pd.read_csv("who.csv")
```

```python
gw=gw.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index('Country Name')
```

```python
akws={'fontsize':10,
     'fontstyle':'italic',
     'color':'k',
     'alpha':0.99,
     'rotation':'vertical',
     'verticalalignment':'center',}
sns.heatmap(gw,vmin=0,vmax=21,cmap='coolwarm',annot=True,annot_kws=akws,linewidth=4,linecolor='k')
```

```python
plt.figure(figsize=(16,9))
ckws={'orientation':'horizontal',
     'shrink':1,
      'extend':'min',
     'extendfrac':0.1
     ,'ticks':np.arange(0,22),
     'drawedges':True}
ax=sns.heatmap(gw,vmin=0,vmax=21,cmap='coolwarm',annot=True,cbar_kws=ckws)
ax.set(title='Heatmap')
sns.set(font_scale=2)
```
