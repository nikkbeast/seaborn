```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gw=pd.read_csv("who.csv")
gw=gw.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index('Country Name')
```

```python
gw.corr()
```

```python
plt.figure(figsize=(16,9))
ax=sns.heatmap(gw.corr(),annot=True,linewidth=3,xticklabels=np.arange(2001,2015),yticklabels=np.arange(2001,2015))
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='k')
plt.title('WHO IS RESPONSIBLE FOR GLOBAL WARMING ?',fontsize=25)
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
can=load_breast_cancer()
can
```

```python
can_df=pd.DataFrame(np.c_[can['data'],can['target']],
                    columns=np.append(can['feature_names'],['target']))
can_df
```

```python
plt.figure(figsize=(25,25))
ax=sns.heatmap(can_df.corr(),annot=True,linewidth=3)
ax.tick_params(size=10,color='w',labelsize=10,labelcolor='k')
plt.title('HEATMAP OF BREAST CANCER PATIENT',fontsize=25)
plt.show()
```
