```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips_df=pd.read_csv("tips.csv")
tips_df
```

```python
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=tips_df,hue='sex',style='sex',palette='hot',dashes=False,
            markers=['o','<'],legend='full')
plt.title("Line Plot")
plt.show()
```

```python
sns.set(style='darkgrid')
sns.lineplot(x='size',y='total_bill',data=tips_df,hue='day',style='day',palette='hot',dashes=False,
            markers=['o','<','*','<'],legend='full')
plt.title("Line Plot")
plt.show()
```
