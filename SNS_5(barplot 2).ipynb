```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips_df=pd.read_csv("tips.csv")
tips_df
```

```python
sns.set()
kw_args={'alpha':0.9,'linestyle':':','linewidth':3,'edgecolor':'r'}
sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,
           **kw_args)
plt.show()
```

```python
kw_args={'alpha':0.9,'linestyle':':','linewidth':3,'edgecolor':'r'}
ax=sns.barplot(x='day',y='total_bill',hue='sex',data=tips_df,
           **kw_args)
ax.set(title="Barplot using seaborn",
       xlabel='Day',
       ylabel='Total Bill')
plt.show()
```
